```python
# specifying new names for the columns

new_cols_dict ={
    'timestamp':'time',
    'cnt':'count',
    't1':'temp_real_C',
    't2':'temp_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

# renaming the column names with new names

bikes.rename(new_cols_dict, axis=1, inplace=True)
```

```python
# normalize the humidity values

bikes.humidity_percent = bikes.humidity_percent / 100
```

```python
# creating dictionary for mapping the values 0-3 to the seasons names
seasonj_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}


# creating a weather dictionary so that we can map the integers to the actual written values

weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}


# mapping the values 0-3 to the actual seasons

bikes.season = bikes.season.map(seasonj_dict)


# mapping the values to the actual written weather

bikes.weather = bikes.weather.map(weather_dict)

bikes.head()
```

```python
# making final dataframe into an excel file

bikes.to_excel('london_bikes_final.xlsx', sheet_name='Data')
```
